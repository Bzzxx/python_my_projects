3.1. –°—Ç–≤–æ—Ä—ñ—Ç—å –∑–≤–æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—é, —è–∫–∞ –ø–æ–∫–∞–∂–µ, —Å–∫—ñ–ª—å–∫–∏ –ø–æ–∫—É–ø–æ–∫ (—ñ–Ω–≤–æ–π—Å—ñ–≤) –±—É–ª–æ –∑—Ä–æ–±–ª–µ–Ω–æ –≤ –∫–æ–∂–Ω—ñ–π –∑ —Ñ—ñ–ª—ñ–π (`Branch`) –∑–∞ —Ä—ñ–∑–Ω–∏–º–∏ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è–º–∏ —Ç–æ–≤–∞—Ä—ñ–≤. –ó–∞–ø–∏—à—ñ—Ç—å —Ç–∞–±–ª–∏—Ü—é –≤ –∑–º—ñ–Ω–Ω—É `invoices_by_category` —ñ –≤–∏–≤–µ–¥—ñ—Ç—å –∑–º—ñ–Ω–Ω—É –Ω–∞ –µ–∫—Ä–∞–Ω.
–¶—è —Ç–∞–±–ª–∏—Ü—è –¥–æ–ø–æ–º–æ–∂–µ –ø—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏, –≤ —è–∫—ñ–π —Ñ—ñ–ª—ñ—ó –∫—É–ø—É—é—Ç—å –Ω–∞–π–±—ñ–ª—å—à–µ —Ç–æ–≤–∞—Ä—ñ–≤ –∫–æ–∂–Ω–æ—ó –∑ –∫–∞—Ç–µ–≥–æ—Ä—ñ–π.

```python
b = invoices_by_category ['Electronic accessories']

top_branch = b.sort_values(ascending=[False]).head(1)
top_branch
```

4-6. **–¢–≤–æ—Ä—á–µ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ —Ä–æ–∑–≤–∏—Ç–æ–∫ –∞–Ω–∞–ª—ñ—Ç–∏—á–Ω–æ–≥–æ –º–∏—Å–ª–µ–Ω–Ω—è**

–ö—Ä–æ–∫ 1. –°—Ñ–æ—Ä–º—É–ª—é–π—Ç–µ –¢–†–ò –ø–∏—Ç–∞–Ω–Ω—è (–≥—ñ–ø–æ—Ç–µ–∑–∏) –¥–æ –Ω–∞—è–≤–Ω–∏—Ö –¥–∞–Ω–∏—Ö, —è–∫—ñ –¥–æ–ø–æ–º–æ–≥–ª–∏ –± –≤–∞–º –∑—Ä–æ–∑—É–º—ñ—Ç–∏, —è–∫—ñ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á—ñ —â–æ, –¥–µ —Ç–∞ –∫–æ–ª–∏ –Ω–∞–π–±—ñ–ª—å—à–µ/–Ω–∞–π–º–µ–Ω—à–µ –∫—É–ø—É—é—Ç—å, –∞–±–∏ –¥–∞—Ç–∏ –Ω–∞ –æ—Å–Ω–æ–≤—ñ —Ü–∏—Ö –≥—ñ–ø–æ—Ç–µ–∑ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü—ñ—ó –±—ñ–∑–Ω–µ—Å—É. –ó–≤—ñ—Å–Ω–æ –ø–∏—Ç–∞–Ω–Ω—è –º–∞—é—Ç—å –±—É—Ç–∏ –Ω–µ —Ç–∏–º–∏, –Ω–∞ —è–∫—ñ –º–∏ –≤–∂–µ –≤—ñ–¥–ø–æ–≤—ñ–ª–∏ –≤ –∑–∞–≤–¥–∞–Ω–Ω—è—Ö –º–æ–¥—É–ª—é.

–ö—Ä–æ–∫ 2. –ó–Ω–∞–π–¥—ñ—Ç—å –≤—ñ–¥–ø–æ–≤—ñ–¥—ñ –Ω–∞ —Å–≤–æ—ó –ø–∏—Ç–∞–Ω–Ω—è –∑ –¥–æ–ø–æ–º–æ–≥–æ—é —Ñ—É–Ω–∫—Ü—ñ–æ–Ω–∞–ª—É pandas.

–ö—Ä–æ–∫ 3. –ù–∞–ø–∏—à—ñ—Ç—å, —è–∫ –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ —Ü–µ –ø–∏—Ç–∞–Ω–Ω—è –º–æ–∂–µ –±—É—Ç–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–∞ –¥–ª—è –ø—Ä–∏–π–Ω—è—Ç—Ç—è –±—ñ–∑–Ω–µ—Å —Ä—ñ—à–µ–Ω—å.   
   
 –ü–∏—Ç–∞–Ω–Ω—è –º–æ–∂—É—Ç—å –±—É—Ç–∏ –±—É–¥—å-—è–∫–æ—ó —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ, –∞–ª–µ —ó—Ö –º–∞—î –±—É—Ç–∏ 3. –ö–æ–∂–Ω–µ –ø–∏—Ç–∞–Ω–Ω—è –æ—Ü—ñ–Ω—é—î—Ç—å—Å—è —è–∫ 1 –∑–∞–≤–¥–∞–Ω–Ω—è. –ë–µ–∑ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ü—å–æ–≥–æ –∑–∞–≤–¥–∞–Ω–Ω—è –î–ó –Ω–µ –ø—Ä–∏–π–º–∞—î—Ç—å—Å—è. –Ø–∫—â–æ —î –ø–∏—Ç–∞–Ω–Ω—è —â–æ–¥–æ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è - –ø–∏—à—ñ—Ç—å —É —á–∞—Ç üôå


```python
# –ê–Ω–∞–ª—ñ–∑ —Ç–∏–ø—É –∫–ª—ñ—î–Ω—Ç—ñ–≤ (Members vs Normal). –Ø–∫—ñ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤, —ñ —è–∫—ñ –≥—Ä—É–ø–∏ –ø–æ–∫—É–ø—Ü—ñ–≤ –∫—É–ø—É—é—Ç—å?
# –©–æ —Ü–µ –¥–∞—î –±—ñ–∑–Ω–µ—Å—É: —Ä–æ–∑—É–º—ñ–Ω–Ω—è, —á–∏ –ø—Ä–æ–≥—Ä–∞–º–∞ –ª–æ—è–ª—å–Ω–æ—Å—Ç—ñ –ø—Ä–∞—Ü—é—î, —è–∫ —ó—ó  –ø–æ–∫—Ä–∞—â–∏—Ç–∏? (–∫—É–¥–∏ –Ω–∞–ø—Ä–∞–≤–∏—Ç–∏ –µ–∫—Å–ª—é–∑–∏–≤–Ω—ñ –ø—Ä–æ–ø–æ–∑–∏—Ü—ñ—ó),
# —Ä–æ–∑–≥–ª—è–Ω—É—Ç–∏ –≤–∞—Ä—ñ–∞–Ω—Ç –ø–æ–∫—Ä–∞—â–µ–Ω–Ω—è –ø—Ä–æ–≥—Ä–∞–º–∏ –ª–æ—è–ª—å–Ω–æ—Å—Ç—ñ (–ø–µ—Ä—Å–æ–Ω–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –ø—Ä–æ–ø–æ–∑–∏—Ü—ñ—ó).
customer_analysis = pd.pivot_table(
    df,
    index=['Customer type', 'Product line'],
    values=['gross income', 'Invoice ID'],
    aggfunc={'gross income': 'sum', 'Invoice ID': 'count'}
)
customer_analysis
```

```python
# C–µ—Ä–µ–¥–Ω—ñ–π —á–µ–∫ –ø–æ —Ñ—ñ–ª—ñ—è—Ö —Ç–∞ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö.
# –Ø–∫—ñ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤ —É —è–∫–∏—Ö —Ñ—ñ–ª—ñ—è—Ö –ø—Ä–æ–¥–∞—é—Ç—å—Å—è –≤ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É –∫—Ä–∞—â–µ –∑–∞ —ñ–Ω—à—ñ?
# –©–æ —Ü–µ –¥–∞—î –±—ñ–∑–Ω–µ—Å—É: —Ä–æ–∑—É–º—ñ–Ω–Ω—è –∫—É–ø—ñ–≤–µ–ª—å–Ω–æ—ó —Å–ø—Ä–æ–º–æ–∂–Ω–æ—Å—Ç—ñ –ø–æ–∫—É–ø—Ü—ñ–≤ –≤ –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –≤—ñ–¥ —Ñ—ñ–ª—ñ—ó,
# –ø–ª–∞–Ω—É–≤–∞–Ω–Ω—è –∞—Å–æ—Ä—Ç–∏–º–µ–Ω—Ç—É (–º–æ–∂–ª–∏–≤–æ, –≤–∞—Ä—Ç–æ –¥–æ–¥–∞—Ç–∏ —Ç–æ–≤–∞—Ä–∏ premium-—Å–µ–≥–º–µ–Ω—Ç—É, —Å–ø–ª–∞–Ω—É–≤–∞—Ç–∏ –∞–∫—Ü—ñ—ó —Ç–∞ –º–∞—Ä–∫–µ—Ç–∏–Ω–≥–æ–≤—ñ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—ñ.

avg_total = pd.pivot_table(
    df,
    index='Branch',
    columns='Product line',
    values='Total',
    aggfunc='mean'
)
avg_total
```

```python
# –ê–Ω–∞–ª—ñ–∑ –º–µ—Ç–æ–¥—ñ–≤ –æ–ø–ª–∞—Ç–∏ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä—ñ—è—Ö.
# –ß–∏ –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—Ç—å —Ä—ñ–∑–Ω—ñ –º–µ—Ç–æ–¥–∏ –æ–ø–ª–∞—Ç–∏ –∑–∞–ª–µ–∂–Ω–æ –≤—ñ–¥ –∫–∞—Ç–µ–≥–æ—Ä—ñ—ó —Ç–æ–≤–∞—Ä—ñ–≤ (–º–æ–∂–ª–∏–≤–æ, –¥–æ—Ä–æ–≥—ñ —Ç–æ–≤–∞—Ä–∏ –∫—É–ø—É—é—Ç—å –∫–∞—Ä—Ç–∫–æ—é, –¥–µ—à–µ–≤—ñ - –≥–æ—Ç—ñ–≤–∫–æ—é)?
# –©–æ —Ü–µ –¥–∞—î –±—ñ–∑–Ω–µ—Å—É: –æ–ø—Ç–∏–º—ñ–∑–∞—Ü—ñ—è —Ä–æ–±–æ—Ç–∏ –∫–∞—Å (—Å–∫—ñ–ª—å–∫–∏ –Ω–æ–≤–∏—Ö —Ç–µ—Ä–º—ñ–Ω–∞–ª—ñ–≤ –ø–æ—Ç—Ä—ñ–±–Ω–æ),
# –ø–∞—Ä—Ç–Ω–µ—Ä—Å—Ç–≤–∞ –∑ –ø–ª–∞—Ç—ñ–∂–Ω–∏–º–∏ —Å–∏—Å—Ç–µ–º–∞–º–∏, —Ä–æ–∑—É–º—ñ–Ω–Ω—è –∫—É–ø—ñ–≤–µ–ª—å–Ω–æ—ó –ø–æ–≤–µ–¥—ñ–Ω–∫–∏ –¥–ª—è —Ä—ñ–∑–Ω–∏—Ö —Å–ø–æ—Å–æ–±—ñ–≤ –æ–ø–ª–∞—Ç–∏.
payment_branch = pd.pivot_table(
    df,
    index=['Branch', 'Payment'],
    columns='Product line',
    values='Invoice ID',
    aggfunc='count'
)
payment_branch
```
