## Завдання 1: Порівняння Pandas vs Matplotlib

**Завдання:**
Побудуйте лінійний графік середньої кількості оренд помісячно впродовж всього періоду в даних двома способами:
1. Використовуючи Pandas (DataFrame.plot())
2. Використовуючи Matplotlib безпосередньо

В обох методах додайте маркери-кружечки. Можна також задати свій відмінний від стандартного колір.

Підказка: отримати потрібний формат даних найзручніше з методом датафрейму `resample`.

**Опишіть свої спостереження:** чим відрізняються 2 побудованих графіки? Який вам більше подобається?

```python
monthly_avg = df['count'].resample('ME').mean()
monthly_avg.head()
```

```python
monthly_avg.plot(
    figsize=(10,5),
    marker='o',
    color='yellow',
    title='Середня кількість оренд по місяцях (Pandas)'
)

plt.xlabel('Дата')
plt.ylabel('Середня кількість оренд')

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    monthly_avg.index,
    monthly_avg.values,
    marker='o',
    color='blue'
)

plt.title('Середня кількість оренд по місяцях (Matplotlib)')
plt.xlabel('Дата')
plt.ylabel('Середня кількість оренд')

plt.grid(True)

plt.show()
```

Графік, побудований за допомогою Pandas, є простішим у реалізації та потребує менше коду. Matplotlib дозволяє більш гнучко налаштовувати графіки, але вимагає більше кроків. Для швидкого аналізу, звісно, зручніше використовувати Pandas, а для більш складних візуалізацій - Matplotlib.

## Завдання 3: Subplot - 2x2 сітка графіків

**Завдання:**
Створіть сітку 2x2 з чотирма різними графіками, використовуючи `plt.subplot()`:
1. Лінійний графік середньої температури помісячно.
2. Стовпчикова діаграма середньої годинної кількості оренд за кварталами.
3. Гістограма вологості за всіма погодинними вимірами.
4. Scatter plot температури vs кількості оренд.

Кожен підграфік має містити всі необхідні підписи. Дашборд має містити назву.

```python
monthly_temp = df.groupby('month')['temp'].mean()

quarter_rentals = df.groupby('season')['count'].mean()
```

```python
plt.figure(figsize=(12,8))

# 1️. Температура по місяцях
plt.subplot(2,2,1)

monthly_temp.plot(marker='o', color='purple')

plt.title('Середня температура по місяцях')
plt.xlabel('Місяць')
plt.ylabel('Температура')
plt.grid(True)


# 2️. Середні оренди по сезонах
plt.subplot(2,2,2)

quarter_rentals.plot(kind='bar', color='skyblue')

plt.title('Середня кількість оренд по сезонах')
plt.xlabel('Сезон')
plt.ylabel('Оренди')
plt.xticks(rotation=0)

# 3️. Гістограма вологості
plt.subplot(2,2,3)

plt.hist(df['humidity'], bins=20, color='brown')

plt.title('Розподіл вологості')
plt.xlabel('Вологість')
plt.ylabel('Кількість')


# 4️. Scatter temp vs count
plt.subplot(2,2,4)

plt.scatter(df['temp'], df['count'], alpha=0.5)

plt.title('Температура vs оренди')
plt.xlabel('Температура')
plt.ylabel('Оренди')


plt.suptitle('Аналіз даних оренди велосипедів')

plt.tight_layout()

plt.show()
```

## Завдання 4: Subplots - об'єктно-орієнтований підхід

**Завдання:**
Створіть той самий набір графіків, але використовуючи `fig, ax = plt.subplots()`.

**Дайте відповідь на питання своїми словами**
- Чим відрізняється підхід побудови кількох графіків на одній фігурі з `plt.subplots()` від `plt.subplot()`?

```python
fig, ax = plt.subplots(2,2, figsize=(12,8))

# 1. Температура по місяцях
ax[0,0].plot(monthly_temp, marker='o', color='red')

ax[0,0].set_title('Середня температура по місяцях')
ax[0,0].set_xlabel('Місяць')
ax[0,0].set_ylabel('Температура')
ax[0,0].grid(True)


# 2️. Оренди по сезонах
ax[0,1].bar(quarter_rentals.index, quarter_rentals.values, color='green')

ax[0,1].set_title('Середня кількість оренд по сезонах')
ax[0,1].set_xlabel('Сезон')
ax[0,1].set_ylabel('Оренди')


# 3️. Гістограма вологості
ax[1,0].hist(df['humidity'], bins=20, color='pink')

ax[1,0].set_title('Розподіл вологості')
ax[1,0].set_xlabel('Вологість')
ax[1,0].set_ylabel('Кількість')


# 4️. Scatter temp vs count
ax[1,1].scatter(df['temp'], df['count'], alpha=0.5, color='teal')

ax[1,1].set_title('Температура vs оренди')
ax[1,1].set_xlabel('Температура')
ax[1,1].set_ylabel('Оренди')


fig.suptitle('Аналіз даних оренди велосипедів')

plt.tight_layout()

plt.show()
```

plt.subplot() — простіший спосіб створення кількох графіків, але менш гнучкий.
plt.subplots() дозволяє керувати кожним графіком окремо через об'єкти axes, тому цей підхід більш зручний для складних графіків.
